## Autism Spectrum Disorder

The number of children diagnosed with Autism Spectrum Disorder (ASD) has been going up over the past decade. It is not clear yet why we see an increase in ASD. One of the challenges is the complexity of ASD identification in children that requires a trained health professional. A better understanding of health factors associated with ASD would be beneficial for simplifying the diagnosis process as well as for learning more about ASD phenomenon and developing new therapies for children.

The goal of this project was to identify health attributes or behavior patterns that are associated with ASD diagnoses. 

Data used in this project came from a telephone survey conducted by Centers for Disease Control and Prevention - National Survey of Childrenâ€™s Health data from 2011-2012 (ftp://ftp.cdc.gov/pub/Health_Statistics/NCHS/slaits/nsch_2011_2012/03_Dataset/nsch1112mimp.zip).

## Data reformating for modeling

```python
# One-hot encoding of categorical features.
# Some of the categorical features have multiple classes, e.g. 1 through 5. We will be recoding these features 
# to 0/1 format.

need_to_be_incoded = [
 'arguing',
 'bullying',
 'unhappy',
 'persistent',
 'challenge_control',
 'curious2',
 'cares_school',
 'homework',
 'relationship',
 'mom_health',
 'dad_health',
 'mom_mentalhealth',
 'dad_mentalhealth',
 'race',
 'mom_education',
 'dad_education',
 'poverty_level'
]

cat_data_encoded = pd.get_dummies(cat_data, prefix=need_to_be_incoded, columns=need_to_be_incoded, drop_first=True)
cat_data_encoded.shape
```

```python
# Continuous features scaling.
scaler = MinMaxScaler(feature_range=(0, 1))
cont_data_scaled = scaler.fit_transform(cont_data)
cont_data_scaled = pd.DataFrame(cont_data_scaled, columns=cont_features1)
cont_data_scaled.shape
```

```python
# Combining categorical and continuous dataframes.
model_data = pd.concat([cat_data_encoded, cont_data_scaled], axis=1, sort=False)
model_data.shape
```

```python
# Splitting data in train and test sets.
X = model_data. drop(['asd'], axis=1)
y = model_data['asd'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# Filling NaN with most common category.
feature_cols3 = X.columns
nan_filler_X = {}
for i in feature_cols3:
    nan_filler_X[i] = X_train[i].dropna().value_counts().index[0]

X_train=X_train.fillna(nan_filler_X)

X_test=X_test.fillna(nan_filler_X)

y_train.fillna(y_train.dropna().value_counts().index[0], inplace = True)
y_train.value_counts(dropna = False)

y_test.fillna(y_test.dropna().value_counts().index[0], inplace = True)

X_test.shape, y_test.shape
```

```python
# Balancing classes. The data is highly imbalanced (2% target class). RandomOverSampler and RandomUnderSampler 
# were tested for data balancing. RandomUnderSampler performed better than RandomOverSampler based on 
# logstic regression f1_score (not shown). Therefore, we are using RandomUnderSampler below.
rus = RandomUnderSampler(random_state=42)
X_train_resampled, y_train_resampled = rus.fit_resample(X_train, y_train)

X_train_resampled.shape, y_train_resampled.shape
```

```python
# Feature selection. LogisticRegression classifier.   
# The feature selection was not used in this analysis. It is not needed due to too few observations present in the data set.

logreg = LogisticRegression(penalty='l1', solver='liblinear')
logreg.fit(X_train_resampled,y_train_resampled)

#Model evaluation: train set.
y_train_pred = logreg.predict(X_train_resampled)
f1_score_train = f1_score(y_train_resampled, y_train_pred)
print(f1_score_train)

confusion_matrix_train = confusion_matrix(y_train_resampled, y_train_pred)
print(confusion_matrix_train)

#Model evaluation: test set.
y_test_pred = logreg.predict(X_test)
f1_score_test = f1_score(y_test, y_test_pred)
print(f1_score_test)

confusion_matrix_test = confusion_matrix(y_test, y_test_pred)
print(confusion_matrix_test)

# Features with non-zero coef.
l1 = pd.DataFrame(logreg.coef_.tolist())
l1 = l1.T
l1['features'] = X_train.columns.tolist()
l1.columns.to_list()
l1_nonzero = l1.loc[l1[0]!=0]
l1_zero = l1.loc[l1[0]==0]

l1_nonzero_list = l1_nonzero['features'].tolist()


# Selecting data with non-zero coef.
nonzero_feature_indexes = []

for i in l1_nonzero_list:
    nonzero_feature = X_train.columns.get_loc(i)
    nonzero_feature_indexes.append(nonzero_feature)

X_train_resampled_selected = X_train_resampled[:, nonzero_feature_indexes]

X_test_selected = X_test.loc[:, l1_nonzero_list]
X_train_resampled.shape, X_train_resampled_selected.shape, X_test_selected.shape
```

## Logistic Regression

```python
## Logistic Regression classifier. Parameter search. All features.

parameters = {'penalty':('l1', 'l2'), 'C':[0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]}
logreg = LogisticRegression(random_state=42, solver='liblinear')

warnings.filterwarnings('ignore') 

clf = GridSearchCV(logreg, parameters, cv=5, scoring='f1')#make_scorer(f1_score))
clf.fit(X_train_resampled, y_train_resampled)
print('Best parameters: ', clf.best_params_)

# Model evaluation: train set.
y_train_pred = clf.predict(X_train_resampled)
f1_score_train = f1_score(y_train_resampled, y_train_pred)
print('F1 score for train set:', f1_score_train)

confusion_matrix_train = confusion_matrix(y_train_resampled, y_train_pred)
print(confusion_matrix_train)

# Model evaluation: test set.
y_test_pred = clf.predict(X_test)
f1_score_test = f1_score(y_test, y_test_pred)
print('F1 score for test set:', f1_score_test)

confusion_matrix_test = confusion_matrix(y_test, y_test_pred)
print('Confusion matrix for test set: ')
sns.heatmap(confusion_matrix_test,center=0, cmap='coolwarm', annot=True)
plt.xlabel('Predicted')
plt.ylabel('True');
```

Optimal parameters for Logistic regression are C = 1, penalty = l1.
f1 score for the training set is 0.9046.
f1 score for the test set is 0.2952.
## Random Forest

```python
## Random Forest classifier. Parameter search. All features.

parameters = {'max_features' : [2, 4, 6, 8, 10], 'max_depth' : [2, 4, 6, 8, 10, 100], 
              'min_samples_split': [2, 5, 10, 100], 'min_samples_leaf': [1, 5, 10, 100]}
forest = RandomForestClassifier(n_estimators=100, random_state=42)
clf = GridSearchCV(forest, parameters, cv=5, scoring='f1')
clf.fit(X_train_resampled, y_train_resampled)
print(clf.best_params_)

clf.fit(X_train_resampled,y_train_resampled)

# Model evaluation: train set.
y_train_pred = clf.predict(X_train_resampled)
f1_score_train = f1_score(y_train_resampled, y_train_pred)
print('F1 score for train set:', f1_score_train)

confusion_matrix_train = confusion_matrix(y_train_resampled, y_train_pred)
print('Confusion matrix for train set:')
print(confusion_matrix_train)

# Model evaluation: test set.
y_test_pred = clf.predict(X_test)
f1_score_test = f1_score(y_test, y_test_pred)
print('F1 score for test set:', f1_score_test)

confusion_matrix_test = confusion_matrix(y_test, y_test_pred)
print('Confusion matrix for test set: ')
sns.heatmap(confusion_matrix_test, center=0, cmap='coolwarm', annot=True)
plt.xlabel('Predicted')
plt.ylabel('True');
```

Optimal parameters for Random Forest are max_depth = 8, max_features = 8, min_samples_leaf = 1, min_samples_split = 100.
f1 score for the training set is 0.9080.
f1 score for the test set is 0.2513.

Logistic regression showed better performance (f1 score = 0.2952) than Random Forest (f1 score = 0.2513). This result is most likely driven by lower number of observations present in the data set.
